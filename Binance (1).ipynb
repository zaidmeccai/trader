## TRADER

```python
import requests

market = 'BTCUSDT'
tick_interval = '1h'

url = 'https://api.binance.com/api/v3/klines?symbol='+market+'&interval='+tick_interval
data = requests.get(url).json()

print(type(data))

```

```python
import pandas as pd
import mplfinance as mpf
df=pd.DataFrame(data,columns=['Opentime','Open','High','Low','Close','Volume','Close time','Quote Asset volume','Number of trades','Taker buy base asset volume','Taker buy quote asset volume','Ignore'])
df['Opentime']=pd.to_datetime(df['Opentime'],unit= 'ms')
df.set_index('Opentime',inplace=True)
df=df.astype(float)
mpf.plot(df, type='line', volume=True, figratio=(25,7), style='yahoo', title=market)
```
